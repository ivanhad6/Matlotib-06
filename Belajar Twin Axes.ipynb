```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Kasus Twin Axes 1

x=np.linspace(1,10,25) #2 parameter pertama akan menentukan nilai terkecil dan terbesar yang akan dibentuk dari pemanggilan fungsi linspace
#Parameter ke3 digunakan untuk menentukan jumlah data Point yang akan dihasilkan dari fungsi linspace
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'bs-', label='exp') #'bs' akan membuat warna biru dan marker yang berupa Square dan '-' mengindikasikan adanya garis 
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx() #Berfungsi menghasilkan akses baru yang akan ditempatkan di atas axes yang pertama
ax2.plot(x, np.log(x), 'ro-', label='log') #'ro' akan membuat warna merah dan marker yang berupa bulatan dan '-' mengindikasikan adanya garis 
ax1.set_ylabel('log')

fig.suptitle('Contoh Twin Axes') #Berrfungsi membuat judul
fig.legend(loc='upper left') #Mengatur posisi dari legendnya

plt.show()
```

```python
#Kasus Twin Axes 2

x=np.random.randint(16,40, size=30) #Parameter pertama akan menentukan nilai terkecilnya
                                    #Parameter kedua akan menentukan nilai terbesarnya
                                    #Parameter ke3 akan menentukan size atau ukuran atau jumlah data Point yang akan dihasilkan oleh fungsi randint
x
```

```python
def C2F(celcius=0):
  return(celcius*1.8) + 32

def konversi_sumbu(ax1):
  y1,y2 = ax1.get_ylim() #y1 akan menampung nilai sumbu y terkecil dari axes yang pertama
                         #y2 akan menampung nilai sumbu y terbesar dari axes yang pertama
  ax2.set_ylim(C2F(y1), C2F(y2))
  ax2.figure.canvas.draw()

fig, ax1 = plt.subplots()
ax2 = ax1.twinx() #akan menghasilkan akses baru yang akan ditempatkan di atas axes yang pertama

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperatur Udara') #Untuk membuat judul
plt.show()
```
